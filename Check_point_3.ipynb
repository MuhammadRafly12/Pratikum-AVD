# Data Preparation

Datasen ini didapat dari : https://www.kaggle.com/datasets/abdulszz/spotify-most-streamed-songs

# A. Pengecekan terhadap Data

1. nilai yang hilang (Missing Values)

```python
print((df.isna().sum() / len(df)) * 100)
```

2. Duplikat

```python
df[df.duplicated()]
```

3. Outliers

```python
results = []

cols = df.select_dtypes(include=['float64', 'int64'])

for col in cols:
  q1 = df[col].quantile(0.25)
  q3 = df[col].quantile(0.75)
  iqr = q3 - q1
  lower_bound = q1 - 1.5*iqr
  upper_bound = q3 + 1.5*iqr
  outliers = df[(df[col] < lower_bound) | (df[col] > upper_bound)]
  percent_outliers = (len(outliers)/len(df))*100
  results.append({'Kolom': col, 'Persentase Outliers': percent_outliers})

# Dataframe dari list hasil
results_df = pd.DataFrame(results)
results_df.set_index('Kolom', inplace=True)
results_df = results_df.rename_axis(None, axis=0).rename_axis('Kolom', axis=1)

# Tampilkan dataframe
display(results_df)
```

# B. Penanganan terhadap data


1. Nilai yang hilang (Missing value)

```python
print((df.isna().sum() / len(df)) * 100)
```

2. Melakukan Imputasi

```python
df['in_shazam_charts'].dropna().describe()
```

```python
df['in_shazam_charts'] = df['in_shazam_charts'].astype(str)
```

```python
df['in_shazam_charts'] = df['in_shazam_charts'].str.replace(',', '', regex=False)
```

```python
df['in_shazam_charts'] = pd.to_numeric(df['in_shazam_charts'], errors='coerce')
```

```python
df['in_shazam_charts'] = df['in_shazam_charts'].fillna(df['in_shazam_charts'].dropna().mean())
```

```python
df['key'].dropna().describe()
```

```python
most_common_key = df['key'].mode()[0]
df['key'] = df['key'].fillna(most_common_key)
```

```python
print((df.isna().sum() / len(df)) * 100)
```

3. Duplikat

```python
df.drop_duplicates(inplace=True)
print("\nDuplikat Data setelah Penanganan:")
print(df.duplicated().sum())
```

4. Outliers

```python
columns_to_impute = ['artist_count', 'released_year', 'in_spotify_playlists',
                     'in_spotify_charts', 'in_apple_playlists', 'in_apple_charts',
                     'in_deezer_charts', 'bpm', 'danceability_%', 'energy_%',
                     'instrumentalness_%', 'liveness_%', 'speechiness_%']

for col in columns_to_impute:
    Q1 = df[col].quantile(0.25)
    Q3 = df[col].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR

    # Menggunakan .loc[] agar tidak muncul SettingWithCopyWarning
    df.loc[:, col] = df[col].clip(lower=lower_bound, upper=upper_bound)
```

**Cek ulang**

```python
results = []

cols = df.select_dtypes(include=['float64', 'int64'])

for col in cols:
  q1 = df[col].quantile(0.25)
  q3 = df[col].quantile(0.75)
  iqr = q3 - q1
  lower_bound = q1 - 1.5*iqr
  upper_bound = q3 + 1.5*iqr
  outliers = df[(df[col] < lower_bound) | (df[col] > upper_bound)]
  percent_outliers = (len(outliers)/len(df))*100
  results.append({'Kolom': col, 'Persentase Outliers': percent_outliers})

# Dataframe dari list hasil
results_df = pd.DataFrame(results)
results_df.set_index('Kolom', inplace=True)
results_df = results_df.rename_axis(None, axis=0).rename_axis('Kolom', axis=1)

# Tampilkan dataframe
display(results_df)
```

# **CHECKPOINT 3**

1. Menampilkan musik populer yang paling banyak didengarkan oleh pelanggan

```python
# Mengurutkan data berdasarkan jumlah streams secara menurun dan ambil 10 teratas
top_songs = df.sort_values(by='streams', ascending=False).head(10)
```

```python
plt.figure(figsize=(16, 8))
plt.barh(top_songs['track_name'], top_songs['streams'], color='skyblue')
plt.xlabel('Jumlah Streaming (Spotify)')
plt.ylabel('Judul Lagu')
plt.title('10 Lagu Paling Banyak Didengarkan Berdasarkan Streams di Spotify')
plt.gca().invert_yaxis()  # Membalik urutan agar yang teratas di atas
plt.tight_layout()
```

2. mengidentifikasi genre musik apa yang paling banyak diputar

Dalam konteks musik, "mayor" (major) dan "minor" merujuk pada dua jenis tangga nada (scale) yang berbeda, dengan mayor memberikan kesan ceria dan riang, sedangkan minor memberikan kesan sendu atau sedih

```python
df['streams'] = pd.to_numeric(df['streams'], errors='coerce')

# Mengelompokkan berdasarkan 'key' dan 'mode' lalu menghitung total streams
genre_streams = df.groupby(['key', 'mode'])['streams'].sum().reset_index()

# Membuat kolom baru untuk label genre (kombinasi key dan mode)
genre_streams['genre_label'] = genre_streams['key'] + ' ' + genre_streams['mode']

# Mengurutkan berdasarkan jumlah streams dan ambil 10 teratas
top_genres = genre_streams.sort_values(by='streams', ascending=False).head(10)

# Membuat grafik batang
plt.figure(figsize=(12, 6))
plt.barh(top_genres['genre_label'], top_genres['streams'], color='lightcoral')
plt.xlabel('Total Streams')
plt.ylabel('Genre (Key dan Mode)')
plt.title('10 Genre Musik Paling Banyak Diputar Berdasarkan Streams')
plt.gca().invert_yaxis()
plt.tight_layout()
plt.show()
```

3. Mengidentifikasi artis dan band terfavorit

```python
# Membersihkan kolom 'in_spotify_playlists' (ubah menjadi numerik jika ada nilai non-numerik)
df['in_spotify_playlists'] = pd.to_numeric(df['in_spotify_playlists'], errors='coerce')

# Menghapus baris dengan nilai in_spotify_playlists yang NaN (jika ada)
df = df.dropna(subset=['in_spotify_playlists'])

# Mengelompokkan data berdasarkan artis dan menjumlahkan in_spotify_playlists
artist_playlists = df.groupby('artist(s)_name')['in_spotify_playlists'].sum().reset_index()

# Mengurutkan berdasarkan total in_spotify_playlists (dari besar ke kecil)
top_artists = artist_playlists.sort_values(by='in_spotify_playlists', ascending=False).head(10)

# Membuat visualisasi
plt.figure(figsize=(12, 8))  # Meningkatkan ukuran grafik untuk kejelasan
plt.barh(top_artists['artist(s)_name'], top_artists['in_spotify_playlists'], color='lightgreen')
plt.xlabel('Total Jumlah Playlist di Spotify')
plt.ylabel('Nama Artis/Band')
plt.title('10 Artis/Band Terfavorit Berdasarkan Jumlah Playlist di Spotify')

# Memformat label sumbu x agar lebih mudah dibaca
plt.ticklabel_format(style='plain', axis='x')
plt.gca().xaxis.set_major_formatter(plt.FuncFormatter(lambda x, p: format(int(x), ',')))
plt.gca().invert_yaxis()
plt.tight_layout()
plt.show()

# Menampilkan daftar artis/band terfavorit dalam bentuk teks
print("10 Artis/Band Terfavorit Berdasarkan Jumlah Playlist:")
for idx, row in top_artists.iterrows():
    print(f"{row['artist(s)_name']} - {row['in_spotify_playlists']:,} playlist")
```
